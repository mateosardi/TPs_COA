```python
import math
import matplotlib.pyplot as plt

# Inicialización de variables
x = [0] # La consigna pide que el primer valor sea 0
N = 1000 # 1000 datos

# Generación de la señal secuencia numérica pseudoaleatoria
for i in range(1, N):
    xi = math.pow(math.pi + x[i-1], 5)
    xi_frac = xi - math.floor(xi) # Se queda con la parte fraccionaria
    x.append(xi_frac) # Añade el valor de xi al arreglo de x

# Histograma con 5 bins
plt.hist(x, bins=5)
plt.title("Histograma con 5 bins")
plt.show()

# Histograma con 10 bins
plt.hist(x, bins=10)
plt.title("Histograma con 10 bins")
plt.show()

# Histograma con 20 bins
plt.hist(x, bins=20)
plt.title("Histograma con 20 bins")
plt.show()
```
